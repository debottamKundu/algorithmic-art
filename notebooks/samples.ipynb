```python
import numpy as np
import pandas as pd
from algorithmic_art.rectangles import generate_art
from algorithmic_art.draw import (
    draw_mosaic,
    draw,
    get_movie_colors,
    get_colors_from_heatmap,
)
```

```python
#
df = generate_art(canvas_height=80, canvas_width=80, iterations=20)
df["Color"] = get_colors_from_heatmap(n_colors=len(df))  # uses viridis by default
draw(df)
```

```python
# same plot but with mosaic
# keep canvas sizes small for mosaic
draw_mosaic(df)
```

```python
# using a different color-map
df = generate_art(canvas_height=80, canvas_width=80, iterations=25)
df["Color"] = get_colors_from_heatmap(
    cmap_name="tab10", n_colors=len(df)
)  # uses viridis by default
draw(df)
```

```python
draw_mosaic(df)
```

```python
# use a wes-anderson pallete
# using a different color-map
df = generate_art(canvas_height=80, canvas_width=80, iterations=6)
df["Color"] = get_movie_colors(
    movie_name="Fantastic", n_colors=len(df)
)  # can use partial movie names, check the wesanderson package for more info
draw(df)
```

```python
draw_mosaic(df)
```

```python
# you can also pass in a custom palette into the movie function, not the best api though

# using a different color-map
blade_runner = np.asarray(["#a30502", "#f78b04", "#2b1718", "#153a42", "#027f93"])
df = generate_art(canvas_height=80, canvas_width=80, iterations=6)
df["Color"] = get_movie_colors(
    custom=blade_runner, n_colors=len(df)
)  # can use partial movie names, check the wesanderson package for more info
draw(df)
```

```python
draw_mosaic(df)
```
